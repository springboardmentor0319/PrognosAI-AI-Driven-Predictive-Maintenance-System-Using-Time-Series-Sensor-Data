```python
#show dataset 
import pandas as pd

cols = ['engine_id', 'cycle', 'op1', 'op2', 'op3'] + [f"sensor{i}" for i in range(1,22)]

df = pd.read_csv("../dataset/train_FD001.txt", sep=r"\s+", header=None)
df.columns = cols


```

```python

#Checking data types(dtypes)
#Every column in pandas has a dtype (data type).
#It tells how pandas stores the data internally.
```

```python
df['sensor2'].dtype
#The dtype property shows the internal data type of a column. For example,
# float64 indicates 64-bit floating-point numbers.
```

```python
#dataypes of all columns
df.dtypes
```

```python
#converting data types(astype)
#astype() converts a column from one data type to another when conversion is valid
df['engine_id'] = df['engine_id'].astype('float64')
df['engine_id'].dtype
```

```python
#earlier engine_id was int now it is converted to float64
```

```python
#index data type - shows index type
df.index.dtype
```

```python
#missing data(NaN)
df.isna().sum()
#counts missing values in each column
```

```python
#selecting Missing values 
#isnull() returns True for missing values, allowing us to filter rows containing NaN.
df[df['sensor2'].isnull()]
```
