```python
#manupilating the index 
#we can change the index of the dataframe using set_index() method
df_engine_index = df.set_index("engine_id")
df_engine_index.head()
```

```python
#engine_id becomes the row label
#It is no longer a normal column
#Now rows are grouped visually by engine_id
```

```python
df_engine_index.loc[1] #so it will help in selecting data from here engine_id 1 it is like grouping
```

```python
#conditional selection:
df['engine_id'] == 1 
#returns a boolean series where True indicates rows where engine_id is 1 ie it will be true till row no 192
```

```python
#conditions with iloc
df.loc[df['engine_id'] == 1]
#return only row which belong to engine id 1
```

```python
#combining condition using & (and) . it will return only those rows where both conditions are true
df.loc[(df['engine_id'] == 1) & (df['cycle'] > 50)]
```

```python
df.loc[(df['engine_id'] == 1) | (df['engine_id'] == 2)]
#using OR operator | it will return rows where either condition is true
```

```python
#built in conditional selection method
df.loc[df['engine_id'].isin([1, 2])]
#it will return rows where engine_id is either 1 or 2 using the isin() method .
```

```python
#isnull() check if any value is missing
df.loc[df['sensor1'].isnull()]
```

```python
df.loc[df['sensor1'].notnull()]
```
