```python
import pandas as pd
import numpy as np
```

```python
file_path = "../dataset/train_FD001.txt"

df = pd.read_csv(file_path, sep=r"\s+", header=None)

columns = (
    ["engine_id", "cycle"]
    + [f"op{i}" for i in range(1,4)]
    + [f"sensor{i}" for i in range(1,22)]
)

df.columns = columns
```

```python
max_cycle = df.groupby("engine_id")["cycle"].transform("max")
df["RUL"] = max_cycle - df["cycle"]
```

```python
# Prepare features and target

model_df = df.drop(columns=["engine_id"])

X = model_df.drop(columns=["RUL"])
y = model_df["RUL"]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
import numpy as np

model = GradientBoostingRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)

scores = cross_val_score(
    model,
    X,
    y,
    cv=5,
    scoring="r2"
)

print("R2 scores for each fold:", scores)
print("Mean R2:", np.mean(scores))
print("Std deviation:", np.std(scores))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.plot(range(1,6), scores, marker='o')

plt.xlabel("Fold")
plt.ylabel("R2 Score")
plt.title("Model Stability Across Folds")

plt.grid(True)

plt.show()
```
