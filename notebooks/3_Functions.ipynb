```python
df = pd.read_csv(r"C:\Projects\PrognosAI\dataset\train_FD001.txt", sep=r"\s+", header=None)
```

```python
column_names = ['engine_id', 'cycle'] + \
               [f'op_setting_{i}' for i in range(1, 4)] + \
               [f'sensor_{i}' for i in range(1, 22)]

df.columns = column_names
```

```python
engine_id_mean = df.engine_id.mean()
df.engine_id.map(lambda x: x - engine_id_mean)
```

```python

sensor_1_mean = df['sensor_1'].mean()

def remean_sensor_1(row):
    row['sensor_1'] = row['sensor_1'] - sensor_1_mean
    return row

df.apply(remean_sensor_1, axis='columns')
```
