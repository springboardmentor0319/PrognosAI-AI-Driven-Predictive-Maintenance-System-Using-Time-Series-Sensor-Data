```python
#show dataset 
import pandas as pd

cols = ['engine_id', 'cycle', 'op1', 'op2', 'op3'] + [f"sensor{i}" for i in range(1,22)]

df = pd.read_csv("../dataset/train_FD001.txt", sep=r"\s+", header=None)
df.columns = cols

df.head()
```

```python
sensor1_mean = df['sensor1'].mean()

df['sensor1'].map(lambda x: x - sensor1_mean)
##not understood
```

```python
#apply()
def adjust_sensor1(row): #it will take one full row of the dataframe it will contain all columns engine_id ,cycle etc
    row['sensor1'] = row['sensor1'] - sensor1_mean  #This subtracts the mean from that row’s sensor1 value
    return row #This gives back the updated row.

df.apply(adjust_sensor1, axis='columns')
```

```python
#sensor1_mean = 100
#each value = 100
#100 - 100 = 0
#sensor1 = 0.0 everywhere
# map() works on single coumn and apply() works on whole dataframe
#It returns a new DataFrame and does not modify the original data.
```

```python
#vectorized Opertaion
#Instead of using map() or apply(), we directly subtract the mean from the entire column.
#Pandas automatically applies the operation element-wise.
#Vectorized operations are faster and more efficient than map() or apply().
df['sensor1'] - sensor1_mean
```
