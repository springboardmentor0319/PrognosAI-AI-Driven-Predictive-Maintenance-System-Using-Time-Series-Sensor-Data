```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

time_index = pd.date_range(start="2024-01-01",
                           end="2024-03-31",
                           freq="D")

values = np.random.randn(len(time_index))

plt.figure()
plt.plot(time_index, values)
plt.title("Time Series Plot")
plt.xlabel("Time Frame")
plt.ylabel("Value")
plt.show()
```
