```python

import pandas as pd

cols = ['engine_id', 'cycle', 'op1', 'op2', 'op3'] + [f"sensor{i}" for i in range(1,22)]

df = pd.read_csv("../dataset/train_FD001.txt", sep=r"\s+", header=None)
df.columns = cols
```

```python
df.groupby('engine_id').engine_id.count()
```

```python
df.groupby('engine_id').cycle.max()
```

```python
df.groupby('engine_id').cycle.min()
```

```python
df.groupby('engine_id')['cycle'].agg(['count', 'min', 'max'])
```

```python
df.groupby(['engine_id', 'cycle']).sensor2.mean()
```

```python
grouped = df.groupby(['engine_id']).cycle.count().reset_index()
print(grouped)
```

```python
engine_life = df.groupby('engine_id').cycle.max().reset_index()

engine_life.sort_values(by='cycle', ascending=False)
```
